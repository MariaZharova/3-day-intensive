# 🔎 Знакомство с Python и анализом данных

Главным объектом в аналитике, конечно, являются данные. С точки зрения python файл с информацией помещается в переменную, мы называем её определённым именем (в нвших дальнейших примерах будет data) и далее можем с ней работать при помощи средств программирования. Для компьютера это, фактически, адрес в физической памяти на диске, по которому эти данные хранятся.

Первым делом подключим необходимые библиотеки и "положим" наш файл с данными в переменную data:

```python
# подключение необходимых библиотек
import pandas as pd
```

```python
# считываем данные по ссылке
data = pd.read_csv('https://raw.githubusercontent.com/MariaZharova/Python-for-Data-Analysis/main/Pandas/train.csv', sep = ',')
```

Теперь файл хранится в data, с ним можно выполнять различные действия. Эти "действия" называются ***методами***, пишутся через точку после data и обычно имеют логичное соответствующее название на английском. 

Давайте посмотрим на основные из них:

```python
# метод head показывает заданное количество первых строк (head - голова)
data.head(3)
```

```python
# А метод tail - последних (tail - хвост)
data.tail(4)
```

```python
# посмотрим колонки в нашей файле
data.columns
```

```python
# и их значения
data.values
```

```python
# можно также обращаться к определёным колонкам, пример для Survived
data.Survived.head(4)
# кстати это ещё пример последовательного применения нескольких методов - сначала 
# берём нужную колонку, а затем head() выводит из неё только первые 5 строк
```

```python
# а этот метод покажет размерность файла
data.shape
```

Все вышеперечисленные методы давали нам информацию про исходный файл. 
Теперь применим к нему некоторые преобразования, которые помогут нам в выявлении закономерностей:

```python
# отсортируем по колонке возраста в порядке убывания
# это пример более сложного метода - теперь в скобках можно указать дополнительные параметры
# здесь можем указать колонку, по которой сортируем (возраст Age)
# и порядок возрастания/убывания (ascening: True/False)
data.sort_values('Survived', ascending=True)
```

```python
# можно удалить дубликаты из данных
# и сразу посмотрим на то, изменился ли размер файла
data.drop_duplicates().shape
```

❗<font color = red> Возможно, Вы заметили, что методы для просмотра основной информации о файл пишутся без скобок на конце, в отличие от методов, которые каким-либо образом изменяют его.</font>

```python
# можно отфильтровать по любому условию
data[(data.Pclass == 3) & (data.Sex == 'male')].head()
```

#### 🌟 Что такое фильтрация и как правильно записать условие

Каждое условие необходимо писать в скобках - это может быть == (равно), != (неравно), >, <;

Между скобками можно поставить "и" (&), "или" (|).

Любые данные можно также отрисовать - это поможет нам в выявлении закономерностей

```python
# нарисуем зависимость выжваемости от возраста
# по горизонтальной оси x указываем возраст
# по вертикальной y выживаемость
data.plot.scatter(x='Age', y='Survived')
```
